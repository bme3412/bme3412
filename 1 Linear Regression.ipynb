```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('cost_revenue_clean.csv')
data.head()
```

```python
data.describe()
```

```python
x = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns = ['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y, alpha=0.3)
plt.title('Film Cost vs Local Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0,450000000)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(x,y)
```

Slope Coefficient:

```python
regression.coef_ # theta_1
```

```python
# intercept
regression.intercept_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y, alpha=0.3)
plt.plot(x, regression.predict(x), color='red', linewidth=4, linestyle='--')

plt.title('Film Cost vs Local Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0,450000000)
plt.show()
```

```python
print('R-Square',regression.score(x, y))
```

```python

def fill_the_fridge(amount):
    print('Open door.')
    print('Walk to the store.')
    print('Buy ' + amount + ' cartons of milk.')
    print('Return with milk.')
fill_the_fridge('five')
```

```python
def milk_mission(amount, destination):
    print('Open door.')
    print('Walk to the ' + destination + '.')
    print('Buy ' + amount + ' cartons of milk.')
    print('Return with milk')
milk_mission('twenty', 'department store')
```

```python
milk_mission('twenty', destination='store')
```
