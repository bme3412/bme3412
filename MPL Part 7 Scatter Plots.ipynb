```python
# can use a logscale with scatterplots as well

data = pd.read_csv(r'/Users/BrendanErhard/Desktop/Python/2019-05-31-data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.tight_layout()
plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75)
plt.xscale('log')
plt.yscale('log')

#add title and labels
plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.show()
```

```python
#use ratio of likes and dislikes in the plot, for color of points and sie

# can use a logscale with scatterplots as well

data = pd.read_csv(r'/Users/BrendanErhard/Desktop/Python/2019-05-31-data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.tight_layout()
plt.scatter(view_count, likes, c=ratio, cmap='summer', edgecolor='black', linewidth=1, alpha=0.75)
cbar=plt.colorbar()
cbar.set_label('Like to Dislike ratio')

# add log scale to reduce impact of outlier
plt.xscale('log')
plt.yscale('log')

#add title and labels
plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.show()
```
