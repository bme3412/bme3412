**3.3 Multivariate Data**
+ plotting multivariate data is tricky - no panacea or template - and need to add new dimensions to figure out gradients of change; example, colors or relative area
+ ex: If wanted to add another variable, such as *sex*, one option to be to color the points based on the variable sex
+ ex: if wanted to add a 4th variable, could add size to the dots - but this is tough for humans to distinguish
+ one way to reduce clusster is to add some value of transparency to the individual points
+ rule of thumb: differnt colors are much easier to disntingusih than different size

```python
# create a color variable based on sex
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1

tips['sex_color'] = tips['sex'].apply(recode_sex)

scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(x=tips['total_bill'],y=tips['tip'],
              #set the size of the dots based on the party size
              # we multiple the values by 10, to make the points bigger
              # and to emphasize the difference
              s=tips['size']*10,
              
              # set the color for the sex
              c=tips['sex_color'],
              
              #set the alpha value so points are more transparent
              # this helps with overlapping points
              alpha=0.5)

axes1.set_title('Total Bill vs Tip, Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
scatter_plot.show()
```

**3.4 Seaborn**

+ seaborn builds on matplotlib by providing a higher level interface for statistical graphics
+ the seaborn libray is tightly integrated with Pandas, as ell as nump, scipy, statsmodels; and is built on top of matplotlib

```python
# load seaborn if you have not done so already
import seaborn as sms

tips = sns.load_dataset("tips")
```

```python
# this subplots function, is a shortcut
# for creating separate figure objects
# and to add individual subplots (axes) to the figure
hist, ax = plt.subplots()

# use the distplot function,from seaborn to create out plot
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Historgram with Density Plot')

plt.show() # we still need matplotlib.pyplot to show the figure
```

```python
# the distplot funciton will pot both a histogram and density plot
# if we just want the histogram, can set the kde parameter to false

hist, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde=False)
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
plt.show()
```

**Density Plots**

```python
# density plots are created by drawing a normal distribution centereaed at each data point
# and then smoothing out the ovelapping plots, so the area under the curve is 1

den, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist=False)
ax.set_title('Total Bill Density')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')
plt.show()
```

```python
# you can also use sns.kdeplot if you want a density plot
# this requires deleting "hist =False"

den, ax = plt.subplots()
ax = sns.kdeplot(tips['total_bill'])
ax.set_title('Total Bill Density')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')
plt.show()
```

**Rug Plots**
+ rug plots are a 1-D representation of a variable's distribution
+ they are typically used with other plots to enhance a visualization

```python
hist_den_rug, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug=True)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
plt.show()
```

**Count Plots (Bar Plot)**
+ very similar to histograms
+ but instead of binning values to produce a distribution, bar plots can be used to count discrete variables

```python
count, ax = plt.subplots()
ax = sns.countplot('day', data=tips)
ax.set_title('Count of Days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
plt.show()
```

**Bivariate Data - Scatterplot**
+ there is no explicit function in searbon called scatter
* instead, use the regplot = plots a scatterplot and fits a regression line
+ if set "fit_reg=False", the plot will only show the scatterplot


```python
scatter, ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
plt.show()
```

```python
# lmplot can also create scatterplots
# regplot creates axes; lmplot creates a figure

fig = sns.lmplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
# we can also create a scatterplot that inlcludes a univariate plot on each axis, using [jointplot]
# jointplot does not return axes, so do not need to create a figure with axes on which to place the plot
# instead, the function creates a [JointGrid] object

joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')

# add a title, set font size
# and move the text above the total bill axes

joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

**Pariwise relationships**
+ visualizing all of the pairwise relationships can be performed using [ pairplot]
+ this will plot a scatterplot between each pair of variables, and a histogram for univariate data


```python
fig = sns.pairplot(tips)
```

```python
# a drawback when using pairplot is there is redundant info; the top half of the figure is the same as the bottom half
# can use [pairgrid] to manually assign the plots for top half and bottom half

pair_grid = sns.PairGrid(tips)
# we can use plt.scatter instead of sns.regplot
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.distplot, rug=True)
plt.show()
```

**3.4 Facets**
+ facet = no need to individually subset data, and lay out the axes in a figure
+ the data must be strucutred such that it is tidydata


```python
anscombe_plot = sns.lmplot(x='x', y='y', data=anscombe, fit_reg=False, col='dataset', col_wrap=2)
```

```python
# the col1 parameter is the variable the plot will facet by
# col_wrap creates a figure that has 2 columns
# if col_wrap is not used, all 4 plots will be plotted in the same row
```

```python
# sometimes you need to create a FacetGrid to determine which variable to facet on

#create the FacetGrid
facet = sns.FacetGrid(tips, col='time')

# for each value in time, plot a histogram of the total bill
facet.map(sns.distplot, 'total_bill', rug=True)
plt.show()
```

```python
# the indivudal facets need not be univariate plots
facet = sns.FacetGrid(tips, col='day', hue='sex')
facet = facet.map(plt.scatter, 'total_bill', 'tip')
facet = facet.add_legend()
plt.show()
```

```python
# if you wanted to continue working in seaborn, could crate the same plot using lmplot

fig = sns.lmplot(x='total_bill', y='tip', data=tips, fit_reg=False, hue='sex', col='day')
plt.show()
```

```python
# could also have 1 variable faceted on the x axis
# could have another variable faceted on the y axis

facet = sns.FacetGrid(tips, col='time', row='smoker', hue='sex') 
facet.map(plt.scatter, 'total_bill', 'tip')
plt.show()
```

```python
# if want all the hue elements to overlap
# can use sns.factorplot

facet = sns.factorplot(x='day', y='total_bill', hue='sex', data=tips, row='smoker', col='time', kind='violin')
```

```python
# can change the plot style with the [sns.set_style] function
# typically this function is run just 1x at the top of the code
# the styles that come with seaborn are [darkgrid], [whitegrid], [dark], [white], and [ticks]
```

```python
# initial plot for comparison

fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
plt.show()
```

```python
# set style and plot
sns.set_style('whitegrid')
fix, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
plt.show()
```

```python
# the folowing code shows what all the styles look like
fig = plt.figure()
seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']
for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
        violin.set_title(style)
fig.tight_layout
plt.show()
```
