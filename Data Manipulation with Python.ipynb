## Summary statistics ##

```python
filename = '/Users/BrendanErhard/Desktop/Python/Datacamp/walmart_sales (1).pkl'
sales = pd.read_pickle(filename)
sales.head()
```

```python
# print info for sales
sales.info()
```

```python
# print mean of weekly sales
# print median of weekly sales
sales['weekly_sales'].mean(), sales['weekly_sales'].median()
```

### Summarizing dates ###

```python
sales['date'].max(), sales['date'].min()
```

### Efficient summaries ###

```python
# write a custom IQR range function
def iqr(column):
    return column.quantile(0.75) - column.quantile(0.25)

print(sales['temperature_c'].agg(iqr))
```

```python
# now print the IQR of temp, fuel prie and unemployment
print(sales[["temperature_c", "fuel_price_usd_per_l", "unemployment"]].agg(iqr))
```

```python
# now include numpy and median
import numpy as np
print(sales[["temperature_c", "fuel_price_usd_per_l", "unemployment"]].agg([iqr,np.median]))
```

### Cumulative statistics ###

```python
# create new dataframe which contains sales for dept 1 of store 1
sales_1_1 = sales[(sales['store'] == 1) & (sales['department'] == 1)]
sales_1_1.head()
```

```python
# sort sales_1_1 by date
sales_1_1 = sales_1_1.sort_values(('date'), ascending=True)

# get the cumsum of weeklt sales and add new column
sales_1_1['cum_weekly_sales'] = sales_1_1['weekly_sales'].cumsum()

# get the cumulative max of weekly sales and add new column
sales_1_1['cum_max_sales'] = sales_1_1['weekly_sales'].cummax()

sales_1_1
```

## Counting ##

### Dropping duplicates ###

```python
# remove rows of sales with duplicate pairs of stores and types
store_types = sales.drop_duplicates(subset=['store', 'type'])
store_types.head()
```

```python
# remove rows of sales with duplicate pairs of stores and departments
store_depts = sales.drop_duplicates(subset=['store', 'department'])
store_depts.head()
```

```python
# subset holiday weeks
# drop duplicate dates
holiday_dates = sales[sales['is_holiday']].drop_duplicates(subset='date')
holiday_dates.head()
```

```python
# select date column and print
print(holiday_dates['date'])
```

### Conting Categorical Values ###

```python
# count the # of stores of each type
store_counts = store_types['type'].value_counts()
store_counts
```

```python
# get the proportion of the store types
store_props = store_types['type'].value_counts(normalize=True)
store_props
```

```python
# count the # of departments of each type and sort in descending order
dept_counts_sorted = store_depts['department'].value_counts(sort=True)
dept_counts_sorted
```

```python
# get proportions of departments of each type
# sort in descending order
dept_props_sorted = store_depts['department'].value_counts(sort=True, normalize=True)
dept_counts_sorted
```

## grouped summary stats ##

```python
# using groupby(), figure out the % of sales that occured at each store
# need to separate out into each type of store
sales_all = sales['weekly_sales'].agg(sum)

# subset for type A stores
sales_A = sales[sales['type'] == 'A']['weekly_sales'].agg(sum)

# subset for type B stores
sales_B = sales[sales['type'] == 'B']['weekly_sales'].agg(sum)

# subset for type C stores
sales_C = sales[sales['type'] == 'C']['weekly_sales'].agg(sum)

# get proportion for each type
sales_prop_by_type = [sales_A, sales_B, sales_C] / sales_all
sales_prop_by_type
```

### Calcualtions with groupby() ###

```python
# group by type and calcualte total weekly sales
sales_by_type = sales.groupby('type')['weekly_sales'].agg(sum)

# get proporiton for each type
sales_prop_by_type = sales_by_type / sales_by_type.agg(sum)
sales_prop_by_type
```

```python
# group by is_holdaiy and calculate total weekly sales
sales_by_type_is_holiday = sales.groupby(['type', 'is_holiday'])['weekly_sales'].sum()
sales_by_type_is_holiday
```

### Multiple grouped summaries ###

```python
# for each store type, aggregate weekly_sales
sales_stats = sales.groupby('type')['weekly_sales'].agg([np.min, np.max, np.mean, np.median])
sales_stats
```

```python
# for each store type, aggregate unemployment and fuel price
# get min, max, mean, median
unemp_fuel_stats = sales.groupby('type')[['unemployment', 'fuel_price_usd_per_l']].agg([np.min, np.max, np.mean, np.median])
unemp_fuel_stats
```

## Pivot Tables ##

```python
# pivot for mean weekly_sales for each store type
mean_sales_by_type = sales.pivot_table(values='weekly_sales', index='type')
mean_sales_by_type
```

```python
# pivot for mean and median sales by store type
mean_med_sales_by_type = sales.pivot_table(values='weekly_sales', index='type', aggfunc=[np.mean, np.median])
mean_med_sales_by_type
```

```python
# pivot for mean weekly_sales by store type and holiday
mean_sales_by_type_holiday = sales.pivot_table(values='weekly_sales', index='type', columns='is_holiday')
mean_sales_by_type_holiday
```

### fill in missing values and sum values with pivot tables ###

```python
sales.pivot_table(values='weekly_sales', index='type', columns='department', fill_value=0)
```

```python
# print mean weekly sales by dept and type
# fill missing values with 0
# sum all columns and rows
sales.pivot_table(values='weekly_sales', index='type', columns='department', fill_value=0, margins=True)
```
