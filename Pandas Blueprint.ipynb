## Styling ##

### Building Styles ###

```python
# style functions can either be passed element-wise
    # use 'applymap'
# or column/ row/table-wise
    # use 'apply'
        # for columns, use axis =0
        # for rows, use axis = 1
        # for the entire table at once, use axis=None
    
# start by creating a random dataframe

np.random.seed(24)
df = pd.DataFrame({'A': np.linspace(1, 10, 10)})
df = pd.concat([df, pd.DataFrame(np.random.randn(10, 4), columns=list('BCDE'))],
               axis=1)
df.iloc[3, 3] = np.nan
df.iloc[0, 2] = np.nan
df
```

#### Make negative numbes red, and positive numbers black ####

```python
# need to write a function
def color_negative_red(val):
    color = 'red' if val < 0 else 'black'
    return 'color: %s' % color

# the cell's style depends only on its value
# so need to use applymap

s = df.style.applymap(color_negative_red)
s
```

#### Highlights the max value of a Series in Yellow ####

```python
# write a function
def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]

# this time need to use apply since it is columnt/table-wise
df.style.apply(highlight_max)
```

#### method chain value-based color and highlight max ####

```python
df.style.applymap(color_negative_red).apply(highlight_max)
```

#### Highlight combined with subsetting columns ####

```python
df.style.apply(highlight_max, subset=['C', 'D'])
```

#### Color values by color with slicing ####

```python
df.style.applymap(color_negative_red, subset=pd.IndexSlice[1:5, ['B','C','D']])
```

#### Changing Formatting/Display Values ####

```python
# convert all to % with 2 decimal
df.style.format("{:.2%}")
```

```python
# convert all to % with 1 decimal
df.style.format("{:.1%}")
```

```python
# use a dictionary to format specific columns
df.style.format({'B':"{:0<4.0f}", 'D': '{:+.2f}'})
```

#### Built-in styles ####

```python
# highlight nulls
df.style.highlight_null(null_color='green')
```

### creating a MultiIndex/Hierarchal Index###

```python
arrays = [['bar','bar', 'baz', 'baz', 'foo', 'foo', 'qux', 'qux'],
          ['one', 'two', 'one', 'two', 'one', 'two', 'one', 'two']]

tuples = list(zip(*arrays))
tuples
```

```python
index = pd.MultiIndex.from_tuples(tuples, names=['first', 'second'])
index
```

```python
s = pd.Series(np.random.randn(8), index=index)
s
```

```python
# can use the MUltiIndex from_product
iterables = [['bat', 'baz','foo', 'qux'], ['one', 'two']]
pd.MultiIndex.from_product(iterables, names=['first', 'second'])
```

```python
# can also construct a MultiIndex from a DataFrame
df = pd.DataFrame([['bar', 'one'], ['bar', 'two'],
                 ['foo', 'one'], ['foo', 'two']],
                 columns=['first', 'second'])
pd.MultiIndex.from_frame(df)
```
