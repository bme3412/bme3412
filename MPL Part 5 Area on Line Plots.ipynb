```python
# import the packages
import pandas as pd
import matplotlib.pyplot as plt

# read the csv file
data=pd.read_csv(r'/Users/BrendanErhard/Desktop/Python/mpl 5.csv')

#specify the variables
ages=data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

#plot the variables using ply.plot
plt.plot(ages, dev_salaries, color='#444444',
        linestyle = '--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

#add fill between method
plt.fill_between(ages, py_salaries)

#add legend
plt.legend()

# add titles
plt.title('Median Salary by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')

#plot it
plt.tight_layout()
plt.show()
```

```python
# can even put in additional conditionals, so it only fills when conditional is met
# fill is met when python plot is above overall median salary

# read the csv file
data=pd.read_csv(r'/Users/BrendanErhard/Desktop/Python/mpl 5.csv')

#specify the variables
ages=data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

#plot the variables using ply.plot
plt.plot(ages, dev_salaries, color='#444444',
        linestyle = '--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

#median, pass through to  3rd arugment, "y2"
overall_median = 57287

#add fill between method
plt.fill_between(ages, py_salaries, overall_median, 
                 where=(py_salaries > overall_median),
                 interpolate=True, alpha=0.25)

#add legend
plt.legend()

# add titles
plt.title('Median Salary by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')

#plot it
plt.tight_layout()
plt.show()
```
