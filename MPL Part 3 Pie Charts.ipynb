```python
import matplotlib.pyplot as plt
```

```python
# pie chart, without data
plt.style.use('fivethirtyeight')

plt.title("Pie Chart")
plt.tight_layout()
plt.show()
```

```python
# pie chart, very basic - no labels
plt.style.use('fivethirtyeight')
slices = [60, 40]
plt.pie(slices)

plt.title("Pice Chart")
plt.tight_layout()
plt.show()
```

```python
# pie chart, very basic - now has labels
plt.style.use('fivethirtyeight')
slices = [60, 40]
labels = ['Sixty', 'Forty']

plt.pie(slices, labels=labels)

plt.title("Pie Chart")
plt.tight_layout()
plt.show()
```

```python
# pie chart, very basic - now adding edgecolor to black
plt.style.use('fivethirtyeight')
slices = [60, 40]
labels = ['Sixty', 'Forty']

plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'})

plt.title("Pie Chart")
plt.tight_layout()
plt.show()
```

```python
# pie chart, specfify colors
plt.style.use('fivethirtyeight')
slices = [60, 40, 30, 20]
labels = ['Sixty', 'Forty', 'Extra1', 'Extra2']
colors = ['blue', 'red', 'yellow','green']

plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor': 'black'})

plt.title("Pie Chart")
plt.tight_layout()
plt.show()
```

```python
# pie chart, add hex colors
# pie chart, specfify colors
plt.style.use('fivethirtyeight')
slices = [60, 40, 30, 20]
labels = ['Sixty', 'Forty', 'Extra1', 'Extra2']
colors = ['#008fd5', '#fc4f30', '#e5ae37','#6d904f']

plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor': 'black'})

plt.title("Pie Chart")
plt.tight_layout()
plt.show()
```

```python
#real world example - data from stackoverflow survey

plt.style.use('fivethirtyeight')

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]

labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']

plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'})
plt.title("Programming Languages")
plt.tight_layout()
plt.show()
```

```python
# pie charts don't look good when there is too many data categories
#use a bar chart instead of last video, when have > 5 items
#stripping down data to top 5 langauges

plt.style.use('fivethirtyeight')

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]

labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'})
plt.title("Programming Languages")
plt.tight_layout()
plt.show()
```

```python
# a lot o pie cahrts have an emphasis on 1 piece of the chart; ex: what if you wanted to emphasieepython?
# pass in explode value to offset slice

plt.style.use('fivethirtyeight')

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]

labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

#create an explode list
#0.1 refers to 10% of the radius blowing out
explode = [0, 0, 0, 0.1, 0]

#graph it - make sure to add explode to the parameters
plt.pie(slices, labels=labels, explode=explode, wedgeprops={'edgecolor': 'black'})
plt.title("Programming Languages")
plt.tight_layout()
plt.show()
```

```python
# add more to the blowout/explode ratio --> to 0.5
plt.style.use('fivethirtyeight')

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]

labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

#create an explode list
#0.1 refers to 10% of the radius blowing out
explode = [0, 0, 0, 0.5, 0]

#graph it - make sure to add explode to the parameters
plt.pie(slices, labels=labels, explode=explode, wedgeprops={'edgecolor': 'black'})
plt.title("Programming Languages")
plt.tight_layout()
plt.show()
```

```python
#can add a shadow to the chart

# add more to the blowout/explode ratio --> to 0.5
plt.style.use('fivethirtyeight')

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]

labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

#create an explode list
#0.1 refers to 10% of the radius blowing out
explode = [0, 0, 0, 0.1, 0]

#graph it - make sure to add explode to the parameters
plt.pie(slices, labels=labels, explode=explode, shadow=True, wedgeprops={'edgecolor': 'black'})
plt.title("Programming Languages")
plt.tight_layout()
plt.show()
```

```python
# can also pass in and change the starting angle - this time start angle = 90 -> the result is that it's rotated

plt.style.use('fivethirtyeight')

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

#create an explode list
#0.1 refers to 10% of the radius blowing out
explode = [0, 0, 0, 0.1, 0]

#graph it - make sure to add explode to the parameters
plt.pie(slices, labels=labels, explode=explode, shadow=True, startangle=90, wedgeprops={'edgecolor': 'black'})
plt.title("Programming Languages")
plt.tight_layout()
plt.show()
```

```python
# add in percent labels

plt.style.use('fivethirtyeight')

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

#create an explode list
#0.1 refers to 10% of the radius blowing out
explode = [0, 0, 0, 0.1, 0]

#graph it - make sure to add explode to the parameters
plt.pie(slices, labels=labels, explode=explode, shadow=True,
        startangle=90, autopct='%1.1f%%,
        wedgeprops={'edgecolor': 'black'})
plt.title("Programming Languages")
plt.tight_layout()
plt.show()
```
