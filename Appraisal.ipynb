```python
# libraries
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
# load csv
bub = pd.read_csv(r'/Users/BrendanErhard/Desktop/Bubble.csv')
print(bub)
```

```python
# create data
x = bub['Market Cap']
y = bub['YTD Return']
z = bub['Upside to PT']
```

```python
print(x)
```

```python
# Change color with c and alpha. I map the color to the X axis value.
plt.scatter(x, y, s=x*5, c=x, cmap="Blues", alpha=0.4, edgecolors="grey", linewidth=2)

# Add titles (main and on axis)
plt.xlabel("Market Cap")
plt.ylabel("YTD Return")
plt.title("Upside to PT")

plt.show()
```
