## Back to Back Bar plots in MatplotLib##
source: Chris Albon
https://chrisalbon.com/python/data_visualization/matplotlib_back_to_back_bar_plot/

```python
 # 1. Preliminaries
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('default')
    
```

```python
# 2. create the DataFrame
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
            'pre_score': [4, 24, 31, 2, 3],
            'mid_score': [25, 94, 57, 62,70],
            'post_score': [5, 42, 23, 23, 51]}
df = pd.DataFrame(raw_data, columns=['first_name', 'pre_score', 'mid_score', 'post_score'])
df
```

```python
# 3. Make the Plot - to compare the scores for Molly and Tina
# input data for Molly and Tina
# skip the first column
x1 = df.ix[1, 1:] # all 3 scores for Molly
x2 = df.ix[2, 1:] # all 3 scores for Tina

# create the bar labels
bar_labels = ['Pre Score', 'Mid Score', 'Post Score']

# create a figure
fig = plt.figure(figsize=(9, 6))

# set the y position
y_pos = np.arange(len(x1))
y_pos = [x for x in y_pos]
plt.yticks(y_pos, bar_labels, fontsize=12)

# create a horizontal bar chart in the y position
plt.barh(y_pos,
         # use data from x1
         x1,
         # make it centered
         align ='center',
         # with alpha 0.4
         alpha = 0.4,
         # and color green
         color = '#263F13')

# create a horizontal bar chart in y_pos
plt.barh(y_pos,
         # using NEGATIVE x2
         -x2,
         # make it centered
         align='center',
         #alpha=0.4
         alpha=0.4,
         # and colo green
         color = '#77A61D')

# annotate and add labels
plt.xlabel('Tina\'s Score: Light Green. Molly\'s Score: Dark Green')
t = plt.title('Comparison of Molly and Tina\'s Score')
plt.ylim([-1,len(x1)+0.1])
plt.xlim([-max(x2)-10, max(x1)+10])
plt.grid()

plt.show()
```

```python

# 3. Make the Plot - to compare the scores for Jake and Amy
# input data for Jake and Amy
# skip the first column
x1 = df.ix[3, 1:] # all 3 scores for Jake
x2 = df.ix[4, 1:] # all 3 scores for Amy

# create the bar labels
bar_labels = ['Pre Score', 'Mid Score', 'Post Score']

# create a figure
fig = plt.figure(figsize=(10, 7))

# set the y position
y_pos = np.arange(len(x1))
y_pos = [x for x in y_pos]
plt.yticks(y_pos, bar_labels, fontsize=12)

# create a horizontal bar chart in the y position
plt.barh(y_pos,
         # use data from x1
         x1,
         # make it centered
         align ='center',
         # with alpha 0.4
         alpha = 0.4,
         # and color green
         color = 'blue')

# create a horizontal bar chart in y_pos
plt.barh(y_pos,
         # using NEGATIVE x2
         -x2,
         # make it centered
         align='center',
         #alpha=0.4
         alpha=0.4,
         # and colo green
         color = 'red')

# annotate and add labels
plt.xlabel('Jake\'s Score: Light Green. Amy\'s Score: Dark Green')
t = plt.title('Comparison of Jake and Amy\'s Score')
plt.ylim([-1,len(x1)+0.1])
plt.xlim([-max(x2)-10, max(x1)+10])
plt.grid()

plt.show()
```
